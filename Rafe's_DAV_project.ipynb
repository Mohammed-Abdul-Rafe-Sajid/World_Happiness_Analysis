```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("2015.csv")  # Change filename as per your dataset
df.head()
```

 Checking data types and missing values

```python
df.info()
```

```python
df.dropna(inplace=True)  # Drop missing values if necessary
```

Next, we try to get basic ovrview of data(Statistics)

```python
df.describe()
```

```python

# Rename columns for better readability (if needed)
df.rename(columns={"Happiness.Score": "Happiness_Score"}, inplace=True)
```

```python
# Convert categorical variables to appropriate data types
df["Region"] = df["Region"].astype("category")

```

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
print(df.columns)
```

```python
# Sort data by Happiness Score
df_sorted = df.sort_values(by="Happiness Score", ascending=False)
df_sorted.head()
```

```python
# Sort data bylowest economy
df_sorted = df.sort_values(by="Economy (GDP per Capita)", ascending=True)
df_sorted.head()
```

```python
# Sort data by highest economy
df_sorted = df.sort_values(by="Economy (GDP per Capita)", ascending=False)
df_sorted.head()
```

```python
df["GDP_PER_CAPITA_Rank"] = df["Economy (GDP per Capita)"].rank(ascending=False)
df.head()
```

```python
df = df.drop(columns=["GDP_PER_CAPITA_Rank"])
df.head(4)
```

```python
region_happiness = df.groupby("Region")["Happiness Score"].mean().reset_index()
print(region_happiness)
```

## Exploratory Data Analysis (EDA) & Visualizations

 Histogram of Happiness Scores

```python
plt.figure(figsize=(8, 5))
sns.histplot(df["Happiness Score"], bins=20, kde=True, color="blue")
plt.title("Distribution of Happiness Scores")
plt.xlabel("Happiness Score")
plt.ylabel("Frequency")
plt.show()
```

 Barplot: Happiness Score by Region

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=region_happiness, x="Region", y="Happiness Score", palette="coolwarm")
plt.xticks(rotation=90)
plt.title("Average Happiness Score by Region")
plt.xlabel("Region")
plt.ylabel("Happiness Score")
plt.show()
```

Scatter Plot: GDP vs Happiness

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="Economy (GDP per Capita)", y="Happiness Score", hue="Region", palette="viridis")
plt.title("GDP vs Happiness Score")
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.show()
```

 Boxplot: Happiness Score by Region

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="Region", y="Happiness Score", palette="pastel")
plt.xticks(rotation=90)
plt.title("Happiness Score Distribution by Region")
plt.xlabel("Region")
plt.ylabel("Happiness Score")
plt.show()
```

 Pairplot: Correlation Between Key Features

```python
selected_features = ["Happiness Score", "Economy (GDP per Capita)" , "Freedom", "Generosity"]
sns.pairplot(df[selected_features])
plt.show()
```

Heatmap: Correlation Matrix

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[selected_features].corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Feature Correlation Heatmap")
plt.show()
```
